```python
import numpy as np
import pandas as pd
```

```python
## series of 1-D array
```

```python
s = pd.Series([10,20,30,40])
print(s)
```

```python
pop = pd.Series([1444, 1393, 323, 278],
               index = ["India", "China", "USA", "Indonesia"],
               name = "population_millions"
               )
print(pop)
```

```python
print(pop[pop>500])
```

```python
q1 = pd.Series({"laptops":120, "phones":340, "tablets":77})
q2 = pd.Series({"phones":44, "laptops":474, "watches":73})
print(q1+q2)
```

```python
df = pd.DataFrame({
    "name": ["Asha", "Bill", "Chen", "Diya", "Evan"],
     "dept": ["IT", "HR", "IT", "FINANCE", "HR"],
    "salary": [52000, 48000, 47999, 75000, 50000],
    "city": ["Delhi", "Mumbai", "Surat", "Agra", "Delhi"]
})
print(df)
print(df.shape)
print(list(df.columns))
print(df.head())
df.info()
df.describe()
```

```python
raw = pd.DataFrame({
    "order_id": ["1001", "1002", "1003"],
    "amount": ["2500", "1800", "3200"]
})
print(raw["amount"].sum())
```

```python
clean = raw.copy()
```

```python
clean["order_id"] = pd.to_numeric(clean["order_id"])
clean["amount"] = pd.to_numeric(clean["amount"])
```
