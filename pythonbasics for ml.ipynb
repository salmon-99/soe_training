if-elif grading

```python
marks = float(input("Enter marks: "))

if 90 <= marks <= 100:
    grade = "O"
elif marks >= 80:
    grade = "A+"
elif marks >= 70:
    grade = "A"
elif marks >= 60:
    grade = "B+"
elif marks >= 50:
    grade = "B"
elif marks >= 40:
    grade = "C"
else:
    grade = "F"

print("Grade:", grade)
```

### List Comprehension

```python
numbers = [1,2,3,4]
squares = []
for x in numbers:
    squares.append(x**2)
print(squares)    
```

```python
squares = [x**2 for x in range(5)]
print(squares)
```

```python
even = [x for x in range(20) if x%2 == 0]
print(even)
```

```python
words = ["python", "machine learning", "datascience"]
upper = [w.upper() for w in words]
print(upper)
```
