```python
# missing values
```

```python
messy = pd.DataFrame({
    "product" : ["Pen", "Bag", "Lamp", "Mug", "Book"],
     "price"  : [20, np.nan, 850, 1200, np.nan],
     "rating"  : [4.1, 4.5, np.nan, 8.6, 5.5]
    })
messy
```

```python
messy.isna().sum()
```

```python
filled = messy.copy()
```

```python
filled["price"] = filled["price"].fillna(filled["price"].mean())
```

```python
filled["rating"] = filled["rating"].fillna(filled["rating"].median())
```

```python
print(filled)
```
